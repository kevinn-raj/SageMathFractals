```python
class MyTurtle():
    """
    
    """

    
    def __init__(self, angle = 0, position = vector((0,0))):
        self.__angle = angle
        self.__position = position
    
    def left(self, a):
        self.__angle += a
        
    def right(self, a):
        self.__angle -= a
        
    def forward(self, l):
        d = l * vector((cos(self.__angle), sin(self.__angle)))
        self.__position += d
        
    def back(self, l):
        d = - l * vector((cos(self.__angle), sin(self.__angle)))
        self.__position += d
    
    def get_angle(self):
        return self.__angle
    
    def get_position(self):
        return self.__position
    
    
```

```python
axiom = "F--F--F"
result = axiom
ruleF = "F+F--F+F"

nIterations = 6

for i in range(nIterations):
    result = result.replace("F", ruleF)


t = MyTurtle()
d = 1
a = pi/2

path = [t.get_position()]

for r in result:
    if r == "F":
        t.forward(d)
        p = t.get_position()
        
        path.append(p)
    elif r == "+":
        t.left(a)
    elif r == "-":
        t.right(a)
        
path3d = [vector((p[0], p[1], 0)) for p in path]        
l = line3d(path3d)
show(l, aspect_ratio=1, axes=False, frame=False, viewpoint=([-0.0001,0,1],0.9),
    figsize=(50,50))
```

```python
def grow(axiom, ruleF, angle, distance, iterations = 5):
    result = axiom

    for i in range(iterations):
        tmp = []
        for r in result:
            tmp.append(r.upper().replace("F", ruleF))
        result = "".join(tmp)


    t = MyTurtle()
    d = distance
    a = angle

    path = [t.get_position()]

    for r in result:
        if r.upper() == "F":
            t.forward(d)
            p = t.get_position()

            path.append(p)
        elif r.upper() == "+":
            t.left(a)
        elif r.upper() == "-":
            t.right(a)

    return path

def grow_2(axiom, variables, rules, angle, distance, iterations = 5):
    result = axiom

    for i in range(iterations):
        tmp = []
        for r in result:
            if r == variables[0]:
                tmp.append(r.upper().replace(variables[0], rules[0]))
            elif r == variables[1]:
                tmp.append(r.upper().replace(variables[1], rules[1]))
            else:
                tmp.append(r)
        result = "".join(tmp)

    t = MyTurtle()
    d = distance
    a = angle

    path = [t.get_position()]

    for r in result:
        if r.upper() == variables[0]:
            t.forward(d)
            p = t.get_position()
            path.append(p)
        elif r.upper() == variables[1]:
            t.forward(d)
            p = t.get_position()
            path.append(p)
        elif r.upper() == "+":
            t.left(a)
        elif r.upper() == "-":
            t.right(a)

    return path
```

```python
@interact
def _(distance = 1, angle = slider(0,360,1, 60), 
    iterations = slider(1,10,1,1),
    axiom = input_box("\"F\"", width = 20), rulef = input_box("\"F+F--F+F\"", width = 20, label="rule")):
    
    angle = angle * pi/180
    
    path3d = []
    paths = grow(axiom = str(axiom), ruleF = str(rulef), angle = angle, distance = 1, iterations = iterations)
    
    path3d = [vector((p[0], p[1], 0)) for p in paths]        
    l = line3d(path3d)
    
    show(l, aspect_ratio=1, axes=False, frame=False, viewpoint=([-0.0001,0,1],0.9),
        figsize=(50,50))
    
    "F+F-F-F+F"
```

```python
@interact
def _(angle = slider(0,360,.5, 120), 
    iterations = slider(0,20,1,1),
    axiom = input_box("\"F+G+G\"", width = 20), rulef = input_box("\"F+G-F-G+F\"", width = 20, label="rule 1"),
     ruleg = input_box("\"GG\"", width = 20, label="rule 2")):
    
    angle = angle * pi/180
    
    path3d = []
    paths = grow_2(axiom = str(axiom), variables=("F", "G"), rules = (rulef, ruleg),
                   angle = angle, distance = 1, iterations = iterations)
    
    path3d = [vector((p[0], p[1], 0)) for p in paths]        
    l = line3d(path3d, thickness=1)
    
    show(l, aspect_ratio=1, axes=False, frame=False, viewpoint=([-0.0001,0,1],0.9),
        figsize=(50,50))
    
```
